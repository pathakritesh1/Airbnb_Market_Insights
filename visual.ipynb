```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
abnb = pd.read_excel('D:\\CODE\\Airbnb\\output\\airbnb_cleaned_data.xlsx')
```

```python
# listing price
plt.figure(figsize = (10,6))
sns.histplot(abnb['price'], bins = 30, kde=True)
```

```python
#room type
plt.figure(figsize=(7,5))
sns.countplot(x='room type', data = abnb)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y = 'neighbourhood group', data= abnb,order = abnb['neighbourhood group'].value_counts().index)
```

```python
# relation b/w rooms and price
plt.Figure(figsize=(10,7))
sns.boxplot(x='room type',y='price',hue='room type',data= abnb)
```
